## Loading and Preprocessing your Data

<b>In this section you will use your libraries of choice to load and preprocess your datasets</b>

### Regression Dataset

<b>Please load and preprocess your regression dataset in the cell below.</b>

Note: You may create additional cells in this section to better organize your code.

```python
file_name = 'housing.csv'
dataset = pd.read_csv(file_name)
dataset.info()
dataset.head(5)
```

```python
X = dataset.drop('MEDV', axis=1)
y = dataset['MEDV']
X_train, X_val, y_train, y_val = train_test_split(X, y, random_state=26, test_size=0.2) 
```

### Classification Dataset
<b>Please load and preprocess your classification dataset in the cell below.</b>

Note: You may create additional cells in this section to better organize your code.

```python
file_name2 = 'cardio_train.csv'
dataset2 = pd.read_csv(file_name2, sep=";",index_col=0)
dataset2.info()
dataset2.head(3)
```

```python
X2 = dataset2.drop(['cardio'], axis=1)
y2 = dataset2['cardio']
X_train2, X_val2, y_train2, y_val2 = train_test_split(X2, y2, random_state=26, test_size=0.2)
```

### Regression Model

<b> Please use the cell below to build your regression model.</b>

Note: You may create additional cells in this section to better organize your code.

```python
model_r={} #I used several models, I ll store it in this dictionnary. Final result is below 
```

```python
model = LinearRegression()
model.fit(X_train, y_train)
score = model.score(X_val, y_val)
pred = model.predict([X_train.loc[0]])
print('The model score was {:.4f}'.format(score))
print('The predicted value was {:.2f}'.format(pred[0]))
print('The correct value is {:.2f}'.format(y_train.loc[0]))
model_r.update({'Linear Regression':score})
```

```python
from sklearn.linear_model import Lasso
model = Lasso(alpha=1) #setting the alpha to 1. According to documentation, if we set it to 0, we better go with linearRegression
model.fit(X_train, y_train)
score = model.score(X_val, y_val)
pred = model.predict([X_train.loc[0]])
print('The model score was {:.4f}'.format(score))
print('The predicteCA
9+
d value was {:.2f}'.format(pred[0]))
print('The correct value is {:.2f}'.format(y_train.loc[0]))
model_r.update({'Lasso Regression':score})
```

*Lasso With gridSearch*

```python
alpha = [0.001, 0.01, 0.1, 1, 10, 100, 1000]
param_grid = dict(alpha=alpha)
model = Lasso()
clf=GridSearchCV(model,param_grid)
best_model=clf.fit(X_val,y_val)
```

```python
print ('best alpha:',best_model.best_estimator_.get_params()['alpha'])
print ('score:',best_model.best_estimator_.score(X_val,y_val))
model_r.update({'LR with GridSearch':best_model.best_estimator_.score(X_val,y_val)}) #store the score in the dictionnary
```

### Classification Model

<b> Please use the cell below to build your classification model.</b>

Note: You may create additional cells in this section to better organize your code.

```python
models = {} #dictionnary to store the different scores from models 
```

**Logistic regression**

```python
model = LogisticRegression(max_iter=300)
model.fit(X_train2, y_train2) #Default C=1  solver = lbfgs  penalty = l2
score = model.score(X_val2, y_val2)
pred = model.predict([X_train2.loc[0]])
print('The model score was {}'.format(score))
print('The predicted value was {:.2f}'.format(pred[0]))
print('The correct value is {:.2f}'.format(y_train2.loc[0]))
models.update({'Logistic Regression':score}) #store the score in the dictionnary
```

*GridSearch with logistic Regression*

```python
param_grid = {
    'C': [0.1,1.0, 10.0],
    'penalty': ('l2','l1','none','elasticnet'), 
    'solver':['newton-cg','saga','lbfgs','sag']} 
model = LogisticRegression(max_iter=300)
clf=GridSearchCV(model,param_grid)
best_model=clf.fit(X_val2,y_val2)
```

```python
print ('best C:',best_model.best_estimator_.get_params()['C'])
print ('best penalty:',best_model.best_estimator_.get_params()['penalty'])
print ('best solver:',best_model.best_estimator_.get_params()['solver'])
print ('score:',best_model.best_estimator_.score(X_val2,y_val2))
models.update({'LR with GridSearch':best_model.best_estimator_.score(X_val2,y_val2)}) #store the score in the dictionnary
```

**Gradient Boosting**

```python
model = GradientBoostingClassifier()
model.fit(X_train2, y_train2)
score = model.score(X_val2, y_val2)
pred = model.predict([X_train2.loc[0]])
print('The model score was {}'.format(score))
print('The predicted value was {:.2f}'.format(pred[0]))
print('The correct value is {:.2f}'.format(y_train2.loc[0]))
models.update({'Gradient Boosting':score})
```

*GridSearch with gradient boosting*

```python
#Param_grid choosen based on recommendation article https://www.analyticsvidhya.com/blog/2016/02/complete-guide-parameter-tuning-gradient-boosting-gbm-python/

param_grid = {
    'learning_rate': [0.05,0.1,0.2], #determine the impact on each tree on the final outcome 
    'n_estimators': [50,100,150], #number of sequential tree 
    'subsample':[0.8,0.6,0.9]}, #fraction of observation to be selected in each tree 
model = GradientBoostingClassifier()
clf=GridSearchCV(model,param_grid)
best_model=clf.fit(X_val2,y_val2)
```

```python
print ('best learning rate:',best_model.best_estimator_.get_params()['learning_rate'])
print ('best  n_estimators:',best_model.best_estimator_.get_params()['n_estimators'])
print ('best subsample:',best_model.best_estimator_.get_params()['subsample'])
print ('score:',best_model.best_estimator_.score(X_val2,y_val2))
models.update({'GB with GridSearch':best_model.best_estimator_.score(X_val2,y_val2)})
```

**XGboosting**

```python
from xgboost import XGBClassifier #default: max_depth = 6 ; min child weigth = 1; gamma=0; learning rate =0.3 ; colsample_bytree=1
model = XGBClassifier()
model.fit(X_val2, y_val2)
score = model.score(X_val2, y_val2)
models.update({'XGboost':score})
print(score)
```

*GridSearch with XGboosting*

```python
#Param_grid choosen based on recommendation article https://towardsdatascience.com/doing-xgboost-hyper-parameter-tuning-the-smart-way-part-1-of-2-f6d255a45dde

param_grid = {
    "learning_rate":[0.05, 0.20,0.3] ,
    "max_depth": [ 3, 6, 10],
    "min_child_weight" : [ 1, 3, 5],
    "gamma": [0.1, 0.2],
    "colsample_bytree" : [ 0.3, 0.5 , 1 ] }
model = XGBClassifier()
clf=GridSearchCV(model,param_grid)
best_model=clf.fit(X_val2,y_val2)
```

```python
print ('best learning rate:',best_model.best_estimator_.get_params()['learning_rate'])
print ('max_depth:',best_model.best_estimator_.get_params()['max_depth'])
print ('min_child_weight:',best_model.best_estimator_.get_params()['min_child_weight'])
print ('gamma:',best_model.best_estimator_.get_params()['gamma'])
print ('colsample_bytree:',best_model.best_estimator_.get_params()['colsample_bytree'])
print ('score:',best_model.best_estimator_.score(X_val2,y_val2))
models.update({'XGboost with GridSearch':best_model.best_estimator_.score(X_val2,y_val2)})
```

*randomized search with XGboosting*

```python
param_grid = {
    "learning_rate":[0.05, 0.20,0.3] ,
    "max_depth": [ 3, 6, 10],
    "min_child_weight" : [ 1, 3, 5],
    "gamma": [0.1, 0.2],
    "colsample_bytree" : [ 0.3, 0.5 , 1 ] }
model = XGBClassifier()
clf=RandomizedSearchCV(model,param_grid)
best_model=clf.fit(X_val2,y_val2)
```

```python
print ('best learning rate:',best_model.best_estimator_.get_params()['learning_rate'])
print ('max_depth:',best_model.best_estimator_.get_params()['max_depth'])
print ('min_child_weight:',best_model.best_estimator_.get_params()['min_child_weight'])
print ('gamma:',best_model.best_estimator_.get_params()['gamma'])
print ('colsample_bytree:',best_model.best_estimator_.get_params()['colsample_bytree'])
print ('score:',best_model.best_estimator_.score(X_val2,y_val2))
models.update({'XGboost with RandomSearch':best_model.best_estimator_.score(X_val2,y_val2)})
```
