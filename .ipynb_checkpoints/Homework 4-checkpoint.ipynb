```python
import os
import glob
import pandas as pd
import matplotlib as plt
```

```python
path ='/Users/RoryCochrane/Downloads/homework04_data'
filenames = glob.glob(path + "/*.csv")

dfs = []
for filename in filenames:
    dfs.append(pd.read_csv(filename))

# Concatenate all data into one DataFrame
big_frame = pd.concat(dfs, ignore_index=True)
```

```python
rows2 = big_frame.drop_duplicates()
rows2.head()
```

```python
rows2['monthly income'] = rows2['monthly income'].str.replace('$', '')
rows2['monthly income'] = rows2['monthly income'].astype('float')

rows2.head()
```

```python
med = rows2['age'].median()
```

```python
rows2['age'] = rows2['age'].replace(0,med)

rows2.head()
```

```python
min = min(rows2['age'])
max = max(rows2['age'])

f = lambda x: (x-min)/(max-min)
rows2['age'] = rows2['age'].map(f)

rows2.head()
```

```python
df = rows2.groupby('sex')['ID'].nunique()
df.head(100)
```

```python
rows2['sex'] = rows2['sex'].replace(['M','m','male'],'Male')
rows2['sex'] = rows2['sex'].replace(['F','f','female'],'Female')
rows2['sex'] = rows2['sex'].replace(['0','X'],'Other')

#check that all entries are either male, female or other
df = rows2.groupby('sex')['ID'].nunique()
df.head(100)
```

```python
minw = min(rows2['work hours/week'].astype('int32'))
maxw = max(rows2['work hours/week'].astype('int32'))

f = lambda x: (x-minw)/(maxw-minw)
rows2['work hours/week'] = rows2['work hours/week'].map(f)

rows2.head()
```

```python
rows2['favourite colour'] = rows2['favourite colour'].replace('0','None')
rows2['favourite colour'] = rows2['favourite colour'].replace('BLUE','blue')
rows2['favourite colour'] = rows2['favourite colour'].replace('gReeN','green')
rows2['favourite colour'] = rows2['favourite colour'].replace('reD','red')

df = rows2.groupby('favourite colour')['ID'].nunique()
df.head(100)
```

```python
df = rows2.groupby('siblings')['ID'].nunique()
df.head(100)
```

```python
minsib = rows2['siblings'].astype('int32').min()
maxsib = rows2['siblings'].astype('int32').max()

print(minsib)

f = lambda x: (x-minsib)/(maxsib-minsib)
rows2['sibling'] = rows2['work hours/week'].map(f)

rows2.head()
```

```python
rows2['top finish placement']=rows2['top finish placement'].astype(str).str[0]
rows2['top finish placement']=rows2['top finish placement'].astype('int32')
med = rows2['top finish placement'].median()
rows2['top finish placement'] = rows2['top finish placement'].replace(0,med)

df = rows2.groupby('top finish placement')['ID'].nunique()
df.head(100)
```

```python
df = rows2.groupby('caloric intake')['ID'].nunique()
df.head(100)
```

```python
med = rows2['caloric intake'].median()
rows2['caloric intake'] = rows2['caloric intake'].replace(0,med)

df = rows2.groupby('caloric intake')['ID'].nunique()
df.head(100)
```

```python
rows2['last time'] = rows2['last time'].str.strip('secs')

df = rows2.groupby('last time')['ID'].nunique()
df.head(100)
```

```python
rows2['last time'] = rows2['last time'].astype('int32')
med = rows2['last time'].median()

rows2['last time'] = rows2['last time'].replace(0,med)

df = rows2.groupby('last time')['ID'].nunique()
df.head(100)
```
